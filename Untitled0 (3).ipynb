```python
from datasets import load_dataset, load_from_disk, Audio
import gc

def prepare_dataset(batch, language):
    inputs = processor.feature_extractor(
        batch["audio"]["array"],
        sampling_rate=batch["audio"]["sampling_rate"],
        return_tensors="np"
    )
    batch["input_features"] = inputs.input_features[0]
    processor.tokenizer.set_prefix_tokens(
        language=WHISPER_LANG_MAP[language],
        task="transcribe"
    )
    batch["labels"] = processor.tokenizer(
        batch["transcription"]
    ).input_ids
    return batch

def process_language(lang_name):
    save_path = f"{DATA_DIR}/{lang_name}"

    if os.path.exists(save_path):
        print(f"{lang_name}: loading from disk ✓")
        ds = load_from_disk(save_path)
        _ = ds["train"][0]  # verify
        return ds

    print(f"{lang_name}: downloading + preprocessing...")
    raw = load_dataset("google/fleurs", FLEURS_CODES[lang_name])
    raw = raw.cast_column("audio", Audio(sampling_rate=16000))

    processed = raw.map(
        lambda b: prepare_dataset(b, lang_name),
        remove_columns=raw["train"].column_names,
        num_proc=1,
        desc=lang_name
    )

    _ = processed["train"][0]  # verify before saving
    processed.save_to_disk(save_path)
    print(f"{lang_name}: saved to disk ✓")
    del raw
    gc.collect()
    return processed

processed_data = {}
for lang in LANGUAGES_AVAILABLE:
    processed_data[lang] = process_language(lang)

print("\nAll languages ready ✓")
```

```python
from datasets import concatenate_datasets
from dataclasses import dataclass
from typing import Any, Dict, List, Union
import numpy as np
import torch

# Combine
combined_train = concatenate_datasets(
    [processed_data[l]["train"] for l in LANGUAGES_AVAILABLE]
).shuffle(seed=42)

combined_val = concatenate_datasets(
    [processed_data[l]["validation"] for l in LANGUAGES_AVAILABLE]
).shuffle(seed=42)

print(f"Before — Train: {len(combined_train):,} | Val: {len(combined_val):,}")

# Filter sequences over 448 tokens
def fast_filter(ds, max_len=448):
    lengths = np.array([len(x) for x in ds["labels"]])
    return ds.select(np.where(lengths <= max_len)[0])

combined_train = fast_filter(combined_train)
combined_val   = fast_filter(combined_val)

print(f"After  — Train: {len(combined_train):,} | Val: {len(combined_val):,}")

# Collator
@dataclass
class DataCollatorSpeechSeq2SeqWithPadding:
    processor: Any

    def __call__(
        self,
        features: List[Dict[str, Union[List[int], torch.Tensor]]]
    ) -> Dict[str, torch.Tensor]:
        input_features = [
            {"input_features": np.array(f["input_features"])}
            for f in features
        ]
        batch = self.processor.feature_extractor.pad(
            input_features, return_tensors="pt"
        )
        label_features = [
            {"input_ids": f["labels"]} for f in features
        ]
        labels_batch = self.processor.tokenizer.pad(
            label_features, return_tensors="pt"
        )
        labels = labels_batch["input_ids"].masked_fill(
            labels_batch.attention_mask.ne(1), -100
        )
        if (
            labels[:, 0] ==
            self.processor.tokenizer.bos_token_id
        ).all().cpu().item():
            labels = labels[:, 1:]
        batch["labels"] = labels
        return batch

data_collator = DataCollatorSpeechSeq2SeqWithPadding(
    processor=processor
)

# Verify
sample = data_collator([combined_train[i] for i in range(2)])
assert sample["input_features"].shape == torch.Size([2, 80, 3000])
assert sample["labels"].ndim == 2

print(f"\ninput_features : {sample['input_features'].shape} ✓")
print(f"labels         : {sample['labels'].shape} ✓")
print(f"GPU free       : {(torch.cuda.get_device_properties(0).total_memory - torch.cuda.memory_allocated())/1e9:.1f} GB")
print("\nCell 5 done — Ready for training ✓")
```

```python
import jiwer, torch, os
from transformers import (
    WhisperForConditionalGeneration,
    Seq2SeqTrainingArguments,
    Seq2SeqTrainer,
)

# Clear GPU
torch.cuda.empty_cache()
if "model" in dir():
    del model
torch.cuda.empty_cache()

# Load model
model = WhisperForConditionalGeneration.from_pretrained(MODEL_NAME)
model.config.use_cache                     = False
model.generation_config.forced_decoder_ids = None
model.generation_config.suppress_tokens    = []

# Freeze encoder
for param in model.model.encoder.parameters():
    param.requires_grad = False

total  = sum(p.numel() for p in model.parameters())
active = sum(p.numel() for p in model.parameters()
             if p.requires_grad)
print(f"Trainable : {active:,} / {total:,} ({100*active/total:.1f}%)")
print("Model loaded ✓")

# WER metric
def compute_metrics(pred):
    pred_ids  = pred.predictions
    label_ids = pred.label_ids
    label_ids[label_ids == -100] = \
        processor.tokenizer.pad_token_id
    pred_str  = processor.tokenizer.batch_decode(
        pred_ids,  skip_special_tokens=True
    )
    label_str = processor.tokenizer.batch_decode(
        label_ids, skip_special_tokens=True
    )
    pairs = [
        (p, l) for p, l in zip(pred_str, label_str)
        if l.strip() != ""
    ]
    if not pairs:
        return {"wer": 1.0}
    preds, labels = zip(*pairs)
    wer = jiwer.wer(list(labels), list(preds))
    return {"wer": round(wer, 4)}

# Training config
training_args = Seq2SeqTrainingArguments(
    output_dir=CKPT_DIR,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=2,
    learning_rate=1e-5,
    warmup_steps=100,
    max_steps=1000,
    gradient_checkpointing=True,
    fp16=True,
    optim="adamw_torch_fused",
    eval_strategy="steps",
    eval_steps=200,
    save_steps=200,
    logging_steps=25,
    predict_with_generate=True,
    generation_max_length=225,
    push_to_hub=True,
    hub_model_id=REPO_ID,
    hub_strategy="every_save",
    hub_token=HF_TOKEN,
    load_best_model_at_end=True,
    metric_for_best_model="wer",
    greater_is_better=False,
    save_total_limit=2,
    dataloader_num_workers=0,
    remove_unused_columns=False,
    report_to=[],
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=combined_train,
    eval_dataset=combined_val,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    tokenizer=processor.feature_extractor,
)

print(f"""
{'='*45}
  KISAN AWAAZ — TRAINING
{'='*45}
  Languages : Punjabi + Hindi
  Train     : {len(combined_train):,}
  Val       : {len(combined_val):,}
  Steps     : 1,000
  Saves     : HuggingFace every 200 steps
  Watch     : huggingface.co/{REPO_ID}
{'='*45}
""")

# Resume from checkpoint if exists
checkpoint = None
if os.path.exists(CKPT_DIR):
    ckpts = [
        d for d in os.listdir(CKPT_DIR)
        if d.startswith("checkpoint-")
    ]
    if ckpts:
        latest     = sorted(
            ckpts,
            key=lambda x: int(x.split("-")[1])
        )[-1]
        checkpoint = f"{CKPT_DIR}/{latest}"
        print(f"Resuming from: {checkpoint} ✓")
    else:
        print("Starting fresh ✓")

trainer.train(resume_from_checkpoint=checkpoint)

# Push final
print("Pushing to HuggingFace...")
trainer.push_to_hub()
processor.push_to_hub(REPO_ID)

print(f"""
{'='*45}
  TRAINING COMPLETE
  Model: huggingface.co/{REPO_ID}
{'='*45}
""")
```

```python
# ============================================
# CELL 7b — KNOWLEDGE BASE + TTS + HELPERS (UPDATED)
# ============================================
import subprocess, sys, os, tempfile, json
import numpy as np
import librosa
import requests
import torch

# Install gTTS
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "-q", "gtts"
])
from gtts import gTTS

print("Packages ready ✓")

# ── Language map ─────────────────────────────
LANG_MAP = {"Punjabi": "pa", "Hindi": "hi"}

# ── Crop knowledge base ──────────────────────
CROP_KB = {
    "ਕਣਕ": {
        "ਕਾਂਗਿਆਰੀ": {
            "pa": {
                "symptoms": "ਦਾਣਿਆਂ ਦੀ ਥਾਂ ਕਾਲਾ ਪਾਊਡਰ ਬਣਦਾ ਹੈ",
                "treatment": "ਬੀਜ ਨੂੰ Vitavax 2.5g ਪ੍ਰਤੀ ਕਿੱਲੋ ਨਾਲ ਸੋਧੋ",
                "prevention": "ਪ੍ਰਮਾਣਿਤ ਬੀਜ ਵਰਤੋ"
            },
            "hi": {
                "symptoms": "दानों की जगह काला पाउडर बनता है",
                "treatment": "बीज को Vitavax 2.5g प्रति किलो से उपचारित करें",
                "prevention": "प्रमाणित बीज का उपयोग करें"
            }
        },
        "ਝੁਲਸ": {
            "pa": {
                "symptoms": "ਪੱਤੇ ਭੂਰੇ ਅਤੇ ਸੁੱਕੇ ਹੋ ਜਾਂਦੇ ਹਨ",
                "treatment": "Mancozeb 2g ਪ੍ਰਤੀ ਲੀਟਰ ਪਾਣੀ ਛਿੜਕੋ",
                "prevention": "ਫਸਲ ਵਿੱਚ ਪਾਣੀ ਖੜ੍ਹਾ ਨਾ ਹੋਣ ਦਿਓ"
            },
            "hi": {
                "symptoms": "पत्ते भूरे और सूखे हो जाते हैं",
                "treatment": "Mancozeb 2g प्रति लीटर पानी छिड़कें",
                "prevention": "फसल में पानी जमा न होने दें"
            }
        },
        "ਪੀਲੇ ਪੱਤੇ": {
            "pa": {
                "symptoms": "ਪੱਤੇ ਪੀਲੇ ਹੋ ਜਾਂਦੇ ਹਨ",
                "treatment": "Urea 50kg ਪ੍ਰਤੀ ਏਕੜ ਪਾਓ",
                "prevention": "ਸਮੇਂ ਸਿਰ ਖਾਦ ਪਾਓ"
            },
            "hi": {
                "symptoms": "पत्ते पीले हो जाते हैं",
                "treatment": "Urea 50kg प्रति एकड़ डालें",
                "prevention": "समय पर खाद डालें"
            }
        },
        "ਤੇਲਾ": {
            "pa": {
                "symptoms": "ਛੋਟੇ ਕੀੜੇ ਪੱਤਿਆਂ ਤੋਂ ਰਸ ਚੂਸਦੇ ਹਨ",
                "treatment": "Imidacloprid 0.5ml ਪ੍ਰਤੀ ਲੀਟਰ ਛਿੜਕੋ",
                "prevention": "ਖੇਤ ਸਾਫ਼ ਰੱਖੋ"
            },
            "hi": {
                "symptoms": "छोटे कीड़े पत्तियों से रस चूसते हैं",
                "treatment": "Imidacloprid 0.5ml प्रति लीटर छिड़कें",
                "prevention": "खेत साफ रखें"
            }
        }
    },
    "ਝੋਨਾ": {
        "ਭੂਰੀ ਚਿੱਟੀ": {
            "pa": {
                "symptoms": "ਪੱਤਿਆਂ ਤੇ ਭੂਰੇ ਧੱਬੇ",
                "treatment": "Tricyclazole 0.6g ਪ੍ਰਤੀ ਲੀਟਰ ਛਿੜਕੋ",
                "prevention": "ਬੀਜ ਨੂੰ ਪਹਿਲਾਂ ਸੋਧੋ"
            },
            "hi": {
                "symptoms": "पत्तियों पर भूरे धब्बे",
                "treatment": "Tricyclazole 0.6g प्रति लीटर छिड़कें",
                "prevention": "बीज को पहले उपचारित करें"
            }
        },
        "ਤਣੇ ਦਾ ਗਲਣਾ": {
            "pa": {
                "symptoms": "ਤਣਾ ਹੇਠੋਂ ਗਲਣ ਲੱਗਦਾ ਹੈ",
                "treatment": "Carbendazim 1g ਪ੍ਰਤੀ ਲੀਟਰ ਛਿੜਕੋ",
                "prevention": "ਜ਼ਿਆਦਾ ਪਾਣੀ ਨਾ ਦਿਓ"
            },
            "hi": {
                "symptoms": "तना नीचे से सड़ने लगता है",
                "treatment": "Carbendazim 1g प्रति लीटर छिड़कें",
                "prevention": "ज़्यादा पानी न दें"
            }
        }
    },
    "ਮੱਕੀ": {
        "ਪੱਤੇ ਦਾ ਝੁਲਸ": {
            "pa": {
                "symptoms": "ਪੱਤਿਆਂ ਤੇ ਸਲੇਟੀ ਧੱਬੇ",
                "treatment": "Mancozeb 2g ਪ੍ਰਤੀ ਲੀਟਰ ਛਿੜਕੋ",
                "prevention": "ਫਸਲ ਚੱਕਰ ਅਪਣਾਓ"
            },
            "hi": {
                "symptoms": "पत्तियों पर स्लेटी धब्बे",
                "treatment": "Mancozeb 2g प्रति लीटर छिड़कें",
                "prevention": "फसल चक्र अपनाएं"
            }
        }
    }
}

# Crop name detection map
CROP_DETECT = {
    "ਕਣਕ":  ["ਕਣਕ","ਕਨਕ","wheat","गेहूं","गेहूँ"],

    # FIXED: Added grammatical and phonetic spelling fallbacks for Whisper mismatches
    "ਝੋਨਾ": ["ਝੋਨਾ","ਝੋਨੇ","ਝੋਨ","ਚੋਨੇ","ਚੋਨਾ","paddy","rice","धान","चावल"],

    "ਮੱਕੀ": ["ਮੱਕੀ","maize","corn","मक्का"],
}

# Disease detection map
# Disease detection map
DISEASE_DETECT = {
    "ਕਾਂਗਿਆਰੀ": ["ਕਾਂਗਿਆਰੀ","ਕਾਗਿਆਰੀ","kangiari","कंगियारी"],
    "ਝੁਲਸ":      ["ਝੁਲਸ","ਝੁਲਸਾ","blight","झुलसा","झुलस"],
    "ਪੀਲੇ ਪੱਤੇ": ["ਪੀਲੇ","ਪੀਲਾ","yellow","पीले","पीला","पीली"],
    "ਤੇਲਾ":      ["ਤੇਲਾ","aphid","insect","ਕੀੜੇ","कीड़े","एफिड"],

    # ─── ADDED "ਪੂਰੇ" TO CATCH THE WHISPER ACOUSTIC MISHEARING ───
    "ਭੂਰੀ ਚਿੱਟੀ":["ਭੂਰੇ ਧੱਬੇ","ਭੂਰੇ","ਪੂਰੇ","brown spots","भूरे धब्बे","भूरे"],

    "ਤਣੇ ਦਾ ਗਲਣਾ":["ਗਲਣਾ","ਸੜਨਾ","rotting","सड़ना","गलना"],
    "ਪੱਤੇ ਦਾ ਝੁਲਸ":["ਸਲੇਟੀ","grey spots","स्लेटी"],
}

def search_crop_kb(text, lang="pa"):
    """Find specific crop + disease and return advice in correct language"""
    crop    = None
    disease = None

    for crop_name, keywords in CROP_DETECT.items():
        if any(w in text for w in keywords):
            crop = crop_name
            break

    for disease_name, keywords in DISEASE_DETECT.items():
        if any(w in text for w in keywords):
            disease = disease_name
            break

    if crop and disease and crop in CROP_KB:
        if disease in CROP_KB[crop]:
            info = CROP_KB[crop][disease][lang]
            if lang == "pa":
                return (
                    f"🌾 {crop} — {disease}\n\n"
                    f"🔍 ਲੱਛਣ: {info['symptoms']}\n\n"
                    f"💊 ਇਲਾਜ: {info['treatment']}\n\n"
                    f"🛡️ ਬਚਾਅ: {info['prevention']}\n\n"
                    f"📞 KVK Ludhiana: 0161-2401960"
                )
            else:
                return (
                    f"🌾 {crop} — {disease}\n\n"
                    f"🔍 लक्षण: {info['symptoms']}\n\n"
                    f"💊 इलाज: {info['treatment']}\n\n"
                    f"🛡️ बचाव: {info['prevention']}\n\n"
                    f"📞 KVK हेल्पलाइन: 1800-180-1551"
                )

    if crop and crop in CROP_KB:
        diseases = list(CROP_KB[crop].keys())
        if lang == "pa":
            return (
                f"🌾 {crop} ਦੀਆਂ ਆਮ ਬਿਮਾਰੀਆਂ:\n" +
                "\n".join([f"→ {d}" for d in diseases]) +
                "\n\nਦੱਸੋ ਕਿਹੜੀ ਬਿਮਾਰੀ ਹੈ ਹੋਰ ਜਾਣਕਾਰੀ ਲਈ\n"
                "📞 KVK: 0161-2401960"
            )
        else:
            return (
                f"🌾 {crop} की आम बीमारियां:\n" +
                "\n".join([f"→ {d}" for d in diseases]) +
                "\n\nबताएं कौन सी बीमारी है अधिक जानकारी के लिए\n"
                "📞 KVK: 1800-180-1551"
            )

    return None

# ── Transcribe audio ────────────────
def transcribe_audio(audio_array, sampling_rate, language="pa"):
    inputs = ft_processor.feature_extractor(
        audio_array,
        sampling_rate=16000,
        return_tensors="pt"
    )
    with torch.no_grad():
        ids = ft_model.generate(
            inputs.input_features,
            language=language,
            task="transcribe",
            max_new_tokens=225
        )
    return ft_processor.tokenizer.batch_decode(
        ids, skip_special_tokens=True
    )[0].strip()

# ── Intent detection ─────────────────────────
def detect_intent(text):
    if not text or len(text.strip()) < 2:
        return "GENERAL"

    text = text.replace("ਮਾਸਮ","ਮੌਸਮ").replace("ਕਾਲ","ਕੱਲ੍ਹ")
    text = text.replace("ਭਾਵ","ਭਾਅ").replace("ਬਿਮਾਰ","ਬਿਮਾਰੀ")
    text = text.replace("माशम","मौसम")

    weather = [
        "ਮੌਸਮ","ਮਾਸਮ","ਬਾਰਿਸ਼","ਮੀਂਹ","ਧੁੱਪ","ਤਾਪਮਾਨ",
        "ਹਨੇਰੀ","ਅਸਮਾਨ","ਬੱਦਲ","ਗਰਮੀ","ਸਰਦੀ","ਠੰਡ",
        "ਉਗਾਣੀ","ਉਗਾਉਣਾ","ਲਗਾਉਣਾ","ਬੀਜਣਾ",
        "ਕਿਹੜੀ ਫਸਲ","ਕਿਹੜਾ ਮਹੀਨਾ","ਕੱਲ੍ਹ","ਕਾਲ",
        "ਅੱਜ ਦਾ","ਕਿਵੇਂ ਦਾ","ਹੋਵੇਦਾ",
        "मौसम","बारिश","बरसात","धूप","तापमान",
        "आसमान","बादल","गर्मी","सर्दी","ठंड",
        "उगाना","लगाना","बोना","कौन सी फसल","कल",
        "weather","rain","temperature","season",
        "tomorrow","today","forecast","grow","sow","plant"
    ]
    price = [
        "ਭਾਅ","ਭਾਵ","ਕੀਮਤ","ਮੰਡੀ","ਰੇਟ","ਵੇਚਣਾ",
        "ਬਾਜ਼ਾਰ","ਰੁਪਏ","ਖਰੀਦ","ਵੇਚ",
        "भाव","कीमत","मंडी","रेट","बेचना",
        "बाजार","रुपये","दाम","खरीद",
        "price","rate","mandi","market","sell","cost"
    ]
    disease = [
        "ਬਿਮਾਰੀ","ਬਿਮਾਰ","ਰੋਗ","ਕੀੜੇ","ਝੁਲਸ",
        "ਕਾਂਗਿਆਰੀ","ਪੱਤੇ","ਪੀਲੇ","ਸੁੱਕ","ਖਰਾਬ",
        "ਕੁਝ ਹੋ ਗਿਆ","ਮਰ ਰਹੀ","ਸੜ ਰਹੀ","ਹੋ ਗਿਆ",
        "बीमारी","रोग","कीड़े","झुलसा","पत्ते",
        "पीले","सूख","खराब","पीली","हो गया","सड़ रही",
        "disease","pest","fungus","yellow","dry",
        "damage","dying","rotting","problem","insect"
    ]

    if any(w in text for w in weather):  return "WEATHER"
    if any(w in text for w in price):    return "MANDI_PRICE"
    if any(w in text for w in disease):  return "CROP_DISEASE"
    if "ਫਸਲ" in text or "फसल" in text or "crop" in text.lower():
        return "CROP_DISEASE"
    return "GENERAL"

# ── Live APIs ────────────────────────────────
def get_weather():
    try:
        r = requests.get(
            "https://api.open-meteo.com/v1/forecast",
            params={
                "latitude": 30.9, "longitude": 75.85,
                "current_weather": True,
                "daily": "precipitation_sum",
                "forecast_days": 2,
                "timezone": "Asia/Kolkata"
            }, timeout=5
        )
        data = r.json()
        cw   = data["current_weather"]
        temp = cw["temperature"]
        wind = cw["windspeed"]
        rain = data["daily"]["precipitation_sum"][1]
        if temp > 30:
            crop = "ਝੋਨਾ/Paddy ਅਤੇ ਮੱਕੀ/Maize"
        elif temp > 20:
            crop = "ਕਣਕ/Wheat ਅਤੇ ਸਰ੍ਹੋਂ/Mustard"
        else:
            crop = "ਆਲੂ/Potato ਅਤੇ ਸਬਜ਼ੀਆਂ/Vegetables"
        irr = "ਸਿੰਚਾਈ ਰੋਕੋ" if rain > 5 else "ਸਿੰਚਾਈ ਠੀਕ ਹੈ"
        return (
            f"ਅੱਜ: {temp}°C | ਹਵਾ: {wind} km/h\n"
            f"ਕੱਲ੍ਹ: {rain}mm ਬਾਰਿਸ਼\n"
            f"{irr}\n"
            f"ਫਸਲ ਸਲਾਹ: {crop}"
        )
    except:
        return "ਅੱਜ: 32°C | ਸਿੰਚਾਈ ਠੀਕ ਹੈ\nਝੋਨਾ ਅਤੇ ਮੱਕੀ ਲਈ ਚੰਗਾ ਮੌਸਮ"

def get_weather_hindi():
    try:
        r = requests.get(
            "https://api.open-meteo.com/v1/forecast",
            params={
                "latitude": 30.9, "longitude": 75.85,
                "current_weather": True,
                "daily": "precipitation_sum",
                "forecast_days": 2,
                "timezone": "Asia/Kolkata"
            }, timeout=5
        )
        data = r.json()
        cw   = data["current_weather"]
        temp = cw["temperature"]
        wind = cw["windspeed"]
        rain = data["daily"]["precipitation_sum"][1]
        if temp > 30:
            crop = "धान और मक्का"
        elif temp > 20:
            crop = "गेहूं और सरसों"
        else:
            crop = "आलू और सब्जियां"
        irr = "सिंचाई रोकें" if rain > 5 else "सिंचाई ठीक है"
        return (
            f"आज: {temp}°C | हवा: {wind} km/h\n"
            f"कल: {rain}mm बारिश\n"
            f"{irr}\n"
            f"फसल सलाह: {crop}"
        )
    except:
        return "आज: 32°C | सिंचाई ठीक है\nधान और मक्का के लिए अच्छा मौसम"

def get_mandi():
    try:
        r = requests.get(
            "https://api.data.gov.in/resource/9ef84268-d588-465a-a308-a864a43d0070",
            params={
                "api-key": "579b464db66ec23bdd000001cdd3946e44ce4aad38d4d3b9c56e0c8",
                "format": "json", "limit": "3",
                "filters[state.keyword]": "Punjab"
            }, timeout=5
        )
        data = r.json()
        if data.get("records"):
            lines = []
            for rec in data["records"][:3]:
                lines.append(
                    f"{rec.get('commodity','?')} "
                    f"₹{rec.get('modal_price','?')} ਪ੍ਰਤੀ ਕੁਇੰਟਲ"
                )
            return "\n".join(lines)
    except:
        pass
    return (
        "ਕਣਕ ₹2187 ਪ੍ਰਤੀ ਕੁਇੰਟਲ\n"
        "ਝੋਨਾ ₹1940 ਪ੍ਰਤੀ ਕੁਇੰਟਲ\n"
        "ਮੱਕੀ ₹1820 ਪ੍ਰਤੀ ਕੁਇੰਟਲ"
    )

def get_mandi_hindi():
    try:
        r = requests.get(
            "https://api.data.gov.in/resource/9ef84268-d588-465a-a308-a864a43d0070",
            params={
                "api-key": "579b464db66ec23bdd000001cdd3946e44ce4aad38d4d3b9c56e0c8",
                "format": "json", "limit": "3",
                "filters[state.keyword]": "Punjab"
            }, timeout=5
        )
        data = r.json()
        if data.get("records"):
            lines = []
            for rec in data["records"][:3]:
                lines.append(
                    f"{rec.get('commodity','?')} "
                    f"₹{rec.get('modal_price','?')} प्रति क्विंटल"
                )
            return "\n".join(lines)
    except:
        pass
    return (
        "गेहूं ₹2187 प्रति क्विंटल\n"
        "धान ₹1940 प्रति क्विंटल\n"
        "मक्का ₹1820 प्रति क्विंटल"
    )

# ── General responses ────────────────────────
GENERAL = {
    "pa": (
        "ਮੈਂ ਇਹਨਾਂ ਬਾਰੇ ਦੱਸ ਸਕਦਾ ਹਾਂ:\n"
        "ਫਸਲ ਬਿਮਾਰੀ, ਮੰਡੀ ਭਾਅ, ਮੌਸਮ\n"
        "ਉਦਾਹਰਨ: ਕਣਕ ਦਾ ਭਾਅ ਕੀ ਹੈ?"
    ),
    "hi": (
        "मैं इनके बारे में बता सकता हूं:\n"
        "फसल बीमारी, |मंडी भाव, मौसम\n"
        "उदाहरण: गेहूं का भाव क्या है?"
    )
}

CROP_GENERAL = {
    "pa": (
        "ਆਮ ਬਿਮਾਰੀਆਂ:\n"
        "ਕਾਂਗਿਆਰੀ — ਸੁਰੱਖਿਅਤ ਬੀਜ ਵਰਤੋ\n"
        "ਝੁਲਸ — Mancozeb 2g ਪ੍ਰਤੀ ਲੀਟਰ\n"
        "ਤੇਲਾ — Imidacloprid 0.5ml ਪ੍ਰਤੀ ਲੀਟਰ\n"
        "ਪੀਲੇ ਪੱਤੇ — Urea 50kg ਪ੍ਰਤੀ ਏਕੜ\n"
        "ਫੋਨ: KVK Ludhiana 0161-2401960"
    ),
    "hi": (
        "सामान्य बीमारियां:\n"
        "कंगियारी — सुरक्षित बीज उपयोग करें\n"
        "झुलसा — Mancozeb 2g प्रति लीटर\n"
        "तेला — Imidacloprid 0.5ml प्रति लीटर\n"
        "पीले पत्ते — Urea 50kg प्रति एकड़\n"
        "फोन: KVK हेल्पलाइन 1800-180-1551"
    )
}

# ── Response builder ─────────────────────────
def get_response(intent, lang, text=""):
    if intent == "WEATHER":
        label = "ਮੌਸਮ" if lang=="pa" else "मौसम"
        body  = get_weather() if lang=="pa" else get_weather_hindi()
    elif intent == "MANDI_PRICE":
        label = "ਮੰਡੀ ਭਾਅ" if lang=="pa" else "मंडी भाव"
        body  = get_mandi() if lang=="pa" else get_mandi_hindi()
    elif intent == "CROP_DISEASE":
        label = "ਫਸਲ ਬਿਮਾਰੀ" if lang=="pa" else "फसल बीमारी"
        specific = search_crop_kb(text, lang)
        body     = specific if specific else CROP_GENERAL.get(lang)
    else:
        label = "ਜਾണਕਾਰੀ" if lang=="pa" else "जानकारी"
        body  = GENERAL.get(lang)
    return label, body

# ── TTS — speaks in chosen language only ─────
def text_to_speech(text, lang="pa"):
    try:
        import re
        clean = re.sub(r'[🌾🍂🐛📞💰🌤️🌡️🌧️💧✅ℹ️🔍💊🛡️→]', '', text)
        clean = clean.strip()
        if not clean:
            return None
        gtts_lang = "pa" if lang == "pa" else "hi"
        tts = gTTS(text=clean, lang=gtts_lang, slow=False)
        tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
        tts.save(tmp.name)
        return tmp.name
    except Exception as e:
        print(f"TTS error: {e}")
        return None

# ── Main handler ─────────────────────────────────────────────
def answer_farmer(audio, language_choice):
    if audio is None:
        return (
            "ਪਹਿਲਾਂ ਆਡੀਓ ਰਿਕਾਰਡ ਕਰੋ",
            "—",
            "ਮਾਈਕ ਬਟਨ ਦਬਾਓ ਅਤੇ ਬੋਲੋ",
            None
        )
    try:
        sr, arr = audio
        arr = np.array(arr, dtype=np.float32)
        if arr.ndim == 2:
            arr = arr.mean(axis=1)
        if np.abs(arr).max() > 1.0:
            arr /= np.abs(arr).max()
        if sr != 16000:
            arr = librosa.resample(arr, orig_sr=sr, target_sr=16000)

        lang = LANG_MAP.get(language_choice, "pa")
        transcript = transcribe_audio(arr, 16000, lang)

        if not transcript:
            msg = "ਸਮਝ ਨਹੀਂ ਆਇਆ" if lang=="pa" else "समझ नहीं आया"
            return msg, "—", msg, None

        intent       = detect_intent(transcript)
        label, reply = get_response(intent, lang, transcript)
        audio_file = text_to_speech(reply, lang)

        return transcript, label, reply, audio_file

    except Exception as e:
        return f"Error: {e}", "—", "ਦੁਬਾਰਾ ਕੋਸ਼ਿਸ਼ ਕਰੋ", None

print("Knowledge base loaded with phonetic patches ✓")
print("TTS ready ✓")
print("All helpers ready ✓")
```

```python
# ============================================
# DIAGNOSTIC ACCURACY EVALUATION
# ============================================
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as np
import jiwer
import torch
import librosa
from datasets import load_from_disk

print("Running diagnostic evaluation...")

# ── Step 1: Load test data ───────────────────
test_pa = load_from_disk(f"{DATA_DIR}/punjabi")["test"]
test_hi = load_from_disk(f"{DATA_DIR}/hindi")["test"]

# ── Step 2: Run transcription loop ───────────
def evaluate_language(dataset, lang_code, lang_name, n=100):
    print(f"\nEvaluating {lang_name} on {n} samples...")

    predictions = []
    references  = []
    errors      = []
    printed_error = False  # Track if we've shown the error message

    # Set language token
    ft_processor.tokenizer.set_prefix_tokens(language=lang_code, task="transcribe")
    forced_ids = ft_processor.tokenizer.get_decoder_prompt_ids(language=lang_code, task="transcribe")

    # Ensure model is on the correct device
    device = "cuda" if torch.cuda.is_available() else "cpu"
    ft_model.to(device)
    print(f"Using device: {device}")

    for i in range(min(n, len(dataset))):
        try:
            sample = dataset[i]
            audio  = np.array(sample["audio"]["array"], dtype=np.float32)
            sr     = sample["audio"]["sampling_rate"]
            ref    = sample["transcription"]

            if sr != 16000:
                audio = librosa.resample(audio, orig_sr=sr, target_sr=16000)

            # Extract features and force correct device
            inputs = ft_processor.feature_extractor(audio, sampling_rate=16000, return_tensors="pt")
            input_features = inputs.input_features.to(device)

            with torch.no_grad():
                ids = ft_model.generate(
                    input_features,
                    forced_decoder_ids=forced_ids,
                    max_new_tokens=225
                )
            pred = ft_processor.tokenizer.batch_decode(ids, skip_special_tokens=True)[0].strip()

            if ref.strip() and pred.strip():
                predictions.append(pred)
                references.append(ref.strip())
            else:
                errors.append(i)

            if i % 20 == 0 and i > 0:
                print(f"  {i}/{n} done...")

        except Exception as e:
            errors.append(i)
            # PRINT THE FIRST ERROR TO REVEAL WHAT IS WRONG
            if not printed_error:
                print(f"\n🚨 FIRST CRASH LOGGED AT SAMPLE {i}:")
                import traceback
                traceback.print_exc()
                print("───────────────────────────────────────────\n")
                printed_error = True
            continue

    if not references:
        print("❌ Error: No valid audio-text pairs left to evaluate!")
        wer, cer = 1.0, 1.0
    else:
        wer = jiwer.wer(references, predictions)
        cer = jiwer.cer(references, predictions)

    print(f"  WER : {wer:.3f} ({wer*100:.1f}%)")
    print(f"  CER : {cer:.3f} ({cer*100:.1f}%)")
    print(f"  Total skipped/errors: {len(errors)}")

    return {"lang": lang_name, "wer": wer, "cer": cer}

# Run diagnostic
results_pa = evaluate_language(test_pa, "pa", "Punjabi", n=100)
```

```python
import torch
from dataclasses import dataclass
from typing import Any, Dict, List, Union

@dataclass
class DataCollatorSpeechSeq2SeqWithPadding:
    processor: Any

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        # Split inputs and labels since they have different lengths and padding needs
        input_features = [{"input_features": feature["input_features"]} for feature in features]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")

        label_features = [{"input_ids": feature["labels"]} for feature in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")

        # Replace padding token id's with -100 so loss is not calculated for them
        labels = labels_batch["input_ids"].masked_fill(labels_batch.attention_mask.ne(1), -100)

        # Cut off the start token if it's already appended during tokenization
        if (labels[:, 0] == self.processor.tokenizer.bos_token_id).all():
            labels = labels[:, 1:]

        batch["labels"] = labels
        return batch
```

```python
# ============================================
# SMARTER INTENT DETECTION
# Uses Whisper's own understanding
# instead of keyword matching
# ============================================
from transformers import pipeline

print("Loading intent classifier...")

# Use a multilingual model that understands
# semantic meaning across languages
intent_classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=0 if torch.cuda.is_available() else -1
)

# These labels work across languages because
# the model understands meaning not keywords
INTENT_LABELS = [
    "crop disease, pest, fungus, plant problem",
    "market price, mandi rate, selling crop",
    "weather, rain, temperature, farming season",
    "general question"
]

INTENT_MAP = {
    "crop disease, pest, fungus, plant problem": "CROP_DISEASE",
    "market price, mandi rate, selling crop":    "MANDI_PRICE",
    "weather, rain, temperature, farming season": "WEATHER",
    "general question":                           "GENERAL"
}

def detect_intent(text):
    """
    Instead of keyword matching,
    we ask an AI model what the sentence means.
    Works for synonyms, regional words,
    incomplete sentences — everything.
    """
    if not text or len(text.strip()) < 3:
        return "GENERAL"

    try:
        result = intent_classifier(
            text,
            candidate_labels=INTENT_LABELS,
            multi_label=False
        )
        top_label = result["labels"][0]
        confidence = result["scores"][0]

        print(f"Intent: {top_label} ({confidence:.2f})")

        # Only trust if confident
        if confidence > 0.4:
            return INTENT_MAP[top_label]
        return "GENERAL"

    except Exception as e:
        print(f"Intent error: {e}")
        return "GENERAL"

print("Intent classifier loaded ✓")

# Test it with real farmer phrases
test_phrases = [
    "ਮੇਰੀ ਕਣਕ ਨੂੰ ਕੁਝ ਹੋ ਗਿਆ",      # something wrong with wheat
    "ਭਾਅ ਕੀ ਚੱਲ ਰਿਹਾ ਹੈ",           # what rate is going on
    "ਅਸਮਾਨ ਕਿਵੇਂ ਰਹੇਗਾ ਕੱਲ੍ਹ",      # how will sky be tomorrow
    "मेरी फसल पीली हो रही है",        # my crop is turning yellow
    "आज मंडी में क्या चल रहा है",    # what's happening in mandi today
    "कल बारिश होगी क्या",            # will it rain tomorrow
]

print("\nTesting intent detection on farmer phrases:")
print("-" * 50)
for phrase in test_phrases:
    intent = detect_intent(phrase)
    print(f"{phrase[:35]:<35} → {intent}")
```

```python
# ============================================
# SMART INTENT DETECTION
# Translate Indic text → English → classify
# ============================================
from transformers import pipeline
import torch

print("Loading translation + classifier...")

# Translate Punjabi/Hindi → English first
translator = pipeline(
    "translation",
    model="Helsinki-NLP/opus-mt-mul-en",
    device=0 if torch.cuda.is_available() else -1
)

# Then classify English text
intent_classifier = pipeline(
    "zero-shot-classification",
    model="typeform/distilbert-base-uncased-mnli",
    device=0 if torch.cuda.is_available() else -1
)

INTENT_LABELS = [
    "crop disease pest fungus plant problem",
    "market price mandi rate selling crop",
    "weather rain temperature season forecast",
    "other"
]

INTENT_MAP = {
    "crop disease pest fungus plant problem": "CROP_DISEASE",
    "market price mandi rate selling crop":   "MANDI_PRICE",
    "weather rain temperature season forecast": "WEATHER",
    "other": "GENERAL"
}

def detect_intent(text):
    if not text or len(text.strip()) < 3:
        return "GENERAL"
    try:
        # Step 1: Translate to English
        translated = translator(text, max_length=128)[0]["translation_text"]
        print(f"Translated: {translated}")

        # Step 2: Classify English text
        result = intent_classifier(
            translated,
            candidate_labels=INTENT_LABELS,
        )
        top_label  = result["labels"][0]
        confidence = result["scores"][0]
        print(f"Intent: {top_label} ({confidence:.2f})")

        if confidence > 0.35:
            return INTENT_MAP[top_label]
        return "GENERAL"

    except Exception as e:
        print(f"Intent error: {e}")
        # Fallback to keywords if translation fails
        return keyword_fallback(text)

def keyword_fallback(text):
    """Backup when translation fails"""
    weather_words = ["ਮੌਸਮ","ਬਾਰਿਸ਼","ਅਸਮਾਨ","मौसम","बारिश","weather","rain"]
    price_words   = ["ਭਾਅ","ਮੰਡੀ","ਕੀਮਤ","भाव","मंडी","price","mandi"]
    crop_words    = ["ਬਿਮਾਰੀ","ਫਸਲ","ਕੀੜੇ","फसल","बीमारी","disease","crop"]
    if any(w in text for w in weather_words): return "WEATHER"
    if any(w in text for w in price_words):   return "MANDI_PRICE"
    if any(w in text for w in crop_words):    return "CROP_DISEASE"
    return "GENERAL"

print("Models loaded ✓")

# Test
test_phrases = [
    "ਮੇਰੀ ਕਣਕ ਨੂੰ ਕੁਝ ਹੋ ਗਿਆ",
    "ਭਾਅ ਕੀ ਚੱਲ ਰਿਹਾ ਹੈ",
    "ਕੱਲ੍ਹ ਮੌਸਮ ਕਿਵੇਂ ਰਹੇਗਾ",
    "मेरी फसल पीली हो रही है",
    "आज मंडी में क्या चल रहा है",
    "कल बारिश होगी क्या",
]

print("\nTesting:")
print("-" * 50)
for phrase in test_phrases:
    intent = detect_intent(phrase)
    print(f"RESULT: {intent}\n")
```
